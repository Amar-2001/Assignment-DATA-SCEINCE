```python
# Q1. You are writing code for a company. The requirement of the company is that you create a python
# function that will check whether the password entered by the user is correct or not. The function should
# take the password as input and return the string “Valid Password” if the entered password follows the
# below-given password guidelines else it should return “Invalid Password”.

# Note: 1. The Password should contain at least two uppercase letters and at least two lowercase letters.
# 2. The Password should contain at least a number and three special characters.
# 3. The length of the password should be 10 characters long.


def password(): #definition
    
    pwd=input("Enter the password :")
    num=0
    ucase=0
    lcase=0
    spec=0  

    for i in pwd:
        if i.isupper():
            ucase=ucase+1
        elif i.islower():
            lcase=lcase+1
        elif i.isnumeric():
            num=num+1
        else:
            spec=spec+1
        
    if(ucase>=2 and lcase>=2 and num>=1 and spec>=3 and len(pwd)>=10):
        print("Valid Password")
    else:
        print("Invalid password")

password() # calling
        
    
    
        
```

```python
#  Sort a list of tuples having fruit names and their quantity.
# [("mango",99),("orange",80), ("grapes", 1000)


                  # lambda function
print("List sorted by lambda function ")    
fruits=[("mango",99),("orange",80), ("grapes", 1000)]
sort_fruits=sorted(fruits, key=lambda x: x[1])
print(sort_fruits)  

                   # map function

print("\nList sorted by map function ")    
fruits = [("mango", 99), ("orange", 80), ("grapes", 1000)]

fruit_names = list(map(lambda x: x[0], fruits))
fruit_names.sort()

sorted_fruits = sorted(fruits, key=lambda x: fruit_names.index(x[0]))

print(sorted_fruits)


                    # list comrehension
print("\nBy list comrehension ")  
fruits = [("mango", 99), ("orange", 80), ("grapes", 1000)]

sorted_fruits = [x for x in sorted(fruits, key=lambda x: x[0])]

print(sorted_fruits)




                
            
```

```python
#  Sort a list of integers into positive and negative integers lists.
# [1,2,3,4,5,6,-1,-2,-3,-4,-5,0]

                        # map function
    
print("\nSort postive and negative by map function")
l=[1,2,3,4,5,6,-1,-2,-3,-4,-5,0]
l1=[]
l.sort()
for i in l:
    l1.append(i)# append in l1 in sorted form

sort=list(map(lambda x : x if x>=0 else "",l1))
result_pos=list(filter(lambda x : x if x!="" else x,sort))# to remove ""

sort=list(map(lambda x : x if x<0 else "",l1))
result_neg=list(filter(lambda x : x if x!="" else x,sort))# to remove""

print("positive ",result_pos)
print("negative ",result_neg)




        # filter function
    
   
print("\nSort postive and negative by filter function")
l=[1,2,3,4,5,6,-1,-2,-3,-4,-5,0]
l1=[]
l.sort()
for i in l:
    l1.append(i)# append in l1 in sorted form

sort=list(filter(lambda x : x if x>=0 else "",l1))
result_pos=list(filter(lambda x : x if x!="" else x,sort))# to remove ""

sort=list(filter(lambda x : x if x<0 else "",l1))
result_neg=list(filter(lambda x : x if x!="" else x,sort))# to remove""

print("positive ",result_pos)
print("negative ",result_neg)


              # list comprehension
    
   

print("\nSort postive and negative bylist comprehension ")
l=[1,2,3,4,5,6,-1,-2,-3,-4,-5,0]
l1=[]
l.sort()
for i in l:
    l1.append(i)# append in l1 in sorted form
    
sort=[x if x>0 else False for x in l]
pos_result=list(filter(lambda x : x if x!=False else x,sort))# remove False 
print("Positive number : ",pos_result)

sort=[x if x<0 else False for x in l]
neg_result=list(filter(lambda x : x if x!=False else x,sort))# remove False 
print("Negative number : ",neg_result)






```
